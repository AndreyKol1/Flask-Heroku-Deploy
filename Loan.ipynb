```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("CSV/loan.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Gender'] = df['Gender'].fillna(method='bfill').fillna(method='ffill')
```

```python
df['Gender'].value_counts()
```

```python
df.dropna(subset=['Married'], inplace=True)
```

```python
df.dropna(subset=['Dependents'], inplace=True)
```

```python
df['Dependents'].value_counts()
```

```python
df.loc[df['Dependents'] == '3+', 'Dependents'] = 3
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna(method='bfill').fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
df.loc[df['LoanAmount'].isna()]
```

```python
df.dropna(subset=['LoanAmount'], inplace=True)
```

```python
df.loc[df['Loan_Amount_Term'].isna(), 'Loan_Amount_Term'] = (df['Loan_Amount_Term'].mean()).astype(int)
```

```python
df.dropna(subset=['Credit_History'], inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape[0] / 614
```

```python
df.info()
```

```python
df['Dependents'] = df['Dependents'].astype(int)
```

```python
df = pd.get_dummies(df, columns=['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area'], drop_first=True)
```

```python
df.head()
```

```python
status = {"N" : 0, "Y" : 1}

df['Loan_Status'] = df['Loan_Status'].map(status)
```

```python
df.head()
```

```python
X = df.drop(columns=['Loan_ID', 'Loan_Status'])
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
from lazypredict.Supervised import LazyClassifier

clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_train, X_test, y_train, y_test)

from IPython.display import display
display(models)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score

pipe_lr = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))

pipe_lr.fit(X_train, y_train)
```

```python
cross_val_score(pipe_lr, X_train, y_train, scoring='accuracy', cv=10, n_jobs=-1).mean()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'logisticregression__C' : [0.001, 0.01, 0.1, 1, 10, 100]
}

gs = GridSearchCV(pipe_lr, param_grid, scoring='accuracy', cv=10, n_jobs=-1, refit=True)
```

```python
gs.fit(X_train, y_train)
```

```python
best_lr = gs.best_estimator_
```

```python
cross_val_score(best_lr, X_train, y_train, scoring='accuracy', cv=10, n_jobs=-1).mean()
```

```python
from sklearn.model_selection import learning_curve
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def lr_curve(estimator):
    train_size_abs, train_scores, test_scores = learning_curve(estimator, X_train, y_train, train_sizes=np.linspace(0.1, 1.0, 10), scoring='accuracy', n_jobs=-1)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)

    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.plot(train_size_abs, train_scores_mean, marker='o', color='blue')
    plt.fill_between(train_size_abs, train_scores_mean + train_scores_std, train_scores_mean - train_scores_std, color='blue', alpha=0.5, label='Training accuracy')

    plt.plot(train_size_abs, test_scores_mean, marker='o', color='green')
    plt.fill_between(train_size_abs, test_scores_mean + test_scores_std, test_scores_mean - test_scores_std, color='green', alpha=0.5, label='Validation accuracy')

    plt.grid()
    plt.xlabel('Number of training examples')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.ylim([0.6, 1.03])
    plt.show()
    

def classific_report(estimator):
    y_pred = estimator.predict(X_test)
    
    print(classification_report(y_test, y_pred))
    
def roc_auc(estimator):
    y_pred = estimator.fit(X_train, y_train).predict_proba(X_test)[:, 1]
    
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)
    
    plt.plot(fpr, tpr, label = f'Roc auc score = {auc(fpr, tpr):.2f}', color='gray' )
    
    plt.grid()
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()
    
```

```python
classific_report(pipe_lr)
```

```python
classific_report(best_lr)
```

```python
from sklearn.decomposition import PCA

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)

pca = PCA(n_components=X.shape[1])
principal_components  = pca.fit_transform(scaled_X_train)

explained_variance = pca.explained_variance_ratio_
print(explained_variance)

# Plot explained variance
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1, len(explained_variance) + 1), np.cumsum(explained_variance), where='mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
